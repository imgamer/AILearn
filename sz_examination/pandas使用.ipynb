```python
# 使用数组、字典创建 Series对象
# Series对象是Pandas库中的基石，理解和掌握它是进行高效数据分析的关键。通过Series，你可以轻松地处理【一维数据】，
# 并利用其强大的索引和自动对齐功能来简化数据分析任务。无论是金融分析、时间序列数据处理还是简单的数据统计，Series都是一个非常有用的工具。
#
import pandas as pd

print("使用数组、字典创建 Series对象")
# 使用列表创建对象，索引值为默认的0 1 2... 
se1 = pd.Series([1, 2, 3, 4, 5])
# 使用字典创建Series对象，key值被设置为索引
se2 = pd.Series({"a":1, "b":2, "c":"cccc"})
print(se1)
print()
print(se2)

# 可以通过索引访问元素
#
print("可以通过索引访问元素")
print(se1[0])
print()
# 可以通过位置访问元素，这个访问方式过时了
print(se2[0])
print(se2["c"])

print("\n-------------")
# 修改Label为字符串，否则Lebel为默认的位置
# 索引个数必须和元素个数一致
print("修改Series Label")
se1.index = ['a', 'b', 'c','d', 'e']
print(se1)
print()
print(se2)

# 使用字典把Series对象组合成DataFrame，【注意：2个Series的索引值必须一样】
# key是列名，索引值是合并，没数据会显示为NaN
# NaN 是 "Not a Number" 的缩写，表示一个未定义或不可表示的值
# 在 Python 中，NaN 通常是由 float 类型表示的。你可以使用 float('nan') 来生成一个 NaN 值
print("使用字典把Series对象组合成DataFrame")
df = pd.DataFrame({"se1":se1, "se2":se2})
print(df)
```

```python
# 根据索引访问数据
#
# 用字典创建DF，每个列表是一列，key就是表头表示列名。对应索引的元素组成Series，索引自然0 1 2 ... 相当excel表的行号
#
print("根据索引访问数据")
data = {
    'Artist': ['Billie Holiday', 'Jimi Hendrix', 'Miles Davis', 'SIA'],
    'Genre': ['Jazz', 'Rock', 'Jazz', 'Pop'],
    'Listeners': [1300000, 2700000, 1500000, 2000000],  # 移除了千位分隔符
    'Plays': [27000000, 70000000, 48000000, 74000000]  # 移除了千位分隔符
}
df = pd.DataFrame(data)
print(df)
print()
print(df['Artist'])
print()

print(df[1:3])
```

```python
# 检测缺失数据
#
# 
import pandas as pd
import numpy as np

print("检测和清洗缺失数据:")
# 创建一个示例DataFrame
df = pd.DataFrame({
    'A': [1, np.nan, 3, 4, 5],
    'B': [6, 7, 8, np.nan, 10],
    'C': [11, 12, 13, 14, np.nan]
})
print(df)
# 检测缺失数据
print("\n缺失数据检测:")
print(df.isnull())

# 删除缺失数据
df_dropped = df.dropna()
print("\n删除缺失数据后的DataFrame:")
print(df_dropped)

# 填充缺失数据
df_filled = df.fillna(value = 0)
print("\n填充缺失数据DataFrame:")
print(df_filled)

# 使用前向填充
df_ffilled = df.fillna(method='ffill')
print("\n使用前向填充后的DataFrame:")
print(df_ffilled)

# 使用后向填充
# 有一个nan值后面没有值，填充失败
df_bfilled = df.fillna(method='bfill')
print("\n使用后向填充后的DataFrame:")
print(df_bfilled)

```

```python
# 类型转换
# 底层使用numpy来实现
# 转换时也会检测类型是否满足要求，进行数值范围的合法性检查

import pandas as pd

print("类型转换")
# 创建DataFrame
data_conversion = {
    'StringNumber': ['1', '2', '3', '4'],
    'Date': ['2021-01-01', '2021-02-01', '2021-03-01', '2021-04-01']
}
df_conversion = pd.DataFrame(data_conversion)

print("原始DataFrame:")
print(df_conversion)

# 数值类型转换: 将字符串类型的数字转换为整数
df_conversion['StringNumber'] = df_conversion['StringNumber'].astype(int)
print("\n转换为整数后的DataFrame:")
print(df_conversion)

# 字符串类型转换: 将整数转换回字符串类型
df_conversion['StringNumber'] = df_conversion['StringNumber'].astype(str)
print("\n转换回字符串后的DataFrame:")
print(df_conversion)

# 日期时间类型转换: 将字符串类型的日期转换为日期时间类型
df_conversion['Date'] = pd.to_datetime(df_conversion['Date'])
print("\n转换为日期时间类型后的DataFrame:")
print(df_conversion)
```

```python
print("条件筛选：根据列值筛选")

import pandas as pd
data_filter_sort = {
    'Name': ['John', 'Anna', 'Peter', 'Linda', 'James'],
    'Age': [28, 34, 29, 32, 31],
    'City': ['New York', 'Paris', 'Berlin', 'London', 'Madrid']
}
df_filter_sort = pd.DataFrame(data_filter_sort)

# 使用比较运算符筛选数据：筛选年龄大于30的记录
# 【给DataFrame []传入条件，获得符合条件的所有记录】
#
# 注意这里的语法：一般[]里面是作为索引，但这里面是个【比较表达式，是个条件】，
# 就是说可不仅仅是索引，而是符合条件的记录，多条记录就要用dataframe来放置，条件嵌套了DataFrame的列
df_age_gt_30 = df_filter_sort[df_filter_sort['Age'] > 30]
print("年龄大于30的记录：\n", df_age_gt_30)

# 使用逻辑运算符结合多个条件进行筛选: 筛选年龄大于30且城市为"London"的记录
# 注意，这里修正了城市字段的大小写和变量名的语法
df_age_gt_30_london = df_filter_sort[(df_filter_sort['Age'] > 30) & (df_filter_sort['City'] == 'London')]
print("\n年龄大于30且城市为'London'的记录:\n", df_age_gt_30_london)


# 使用 .isin()方法筛选特定值: 筛选城市为"New York"或"Madrid"的记录
# 注意，这里修正了城市字段的大小写
df_city_specific = df_filter_sort[df_filter_sort['City'].isin(['New York', 'Madrid'])]
print("\n城市为'New York'或'Madrid'的记录:\n", df_city_specific)
```
