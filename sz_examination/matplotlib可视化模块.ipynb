```python
import matplotlib.pyplot as plt

# 数据
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.figure()
plt.plot(x, y)

# 添加标题和标签
plt.title('Simple Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# 显示图形
plt.show()
```

```python
# 同时创建2个子图
#
import matplotlib.pyplot as plt

# 创建一个图形窗口
plt.figure(figsize=(10, 5))

# 第一个子图 (rows, columns, subplot_number)
plt.subplot(1, 2, 1)  # 1行2列的第1个位置
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro-')  # 红色点连线
plt.title('First Subplot')

# 第二个子图
plt.subplot(1, 2, 2)  # 1行2列的第2个位置
plt.plot([1, 2, 3, 4], [1, 2, 3, 4], 'go--')  # 绿色点虚线
plt.title('Second Subplot')

# 使用`plt.subplots_adjust()`调整子图之间的间距
plt.subplots_adjust(wspace=0.4, hspace=0.4)  # 调整子图间的宽度和高度间距

# 显示图形
plt.show()
```

```python
# distplot displot单变量或双变量的分布图
# relplot 两个变量之间的关系图，包括散点图，线图
# catplot 分类数据可视化，支持多种类型的分类图形，如条形图、箱线图
# regplot lmplot 绘制回归分析图，展示变量之间的线性关系
#
# Seaborn
# Seaborn是基于Matplotlib的Python数据可视化库，提供了一种高级接口，用于创建吸引人的统计图形
# Seaborn旨在使可视化成为探索和理解数据的一个核心部分。它内置了许多用于绘制常见统计图形的函数，
# 这些图形不仅外观美观，而且能够展示数据的多维度和结构化特征。
#
# Seaborn的图形创建通常遵循简单的模式：
# 导入Seaborn库，准备数据（通常是Pandas的DataFrame），然后使用一个Seaborn函数来创建一个图形。
# 大多数情况下，你只需提供数据和指定变量即可。
#

# 使用NumPy生成模拟数据集，然后使用Pandas和Seaborn创建一个展示总账单和小费之间线性关系的图表：
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt  # 修正导入语法

# 设置随机种子以确保结果的可重复性
np.random.seed(10)

# 模拟数据集
total_bill = np.random.normal(20, 8, 100)  # 假设账单总额分布，均值为20，标准差为8
tip = total_bill * 0.2 + np.random.normal(2, 1, 100)  # 假设小费与账单总额成正比

# 创建DataFrame
tips = pd.DataFrame({"total_bill": total_bill, "tip": tip})  # 修正变量名

# 创建线性关系图
sns.lmplot(x="total_bill", y="tip", data=tips)  # 修正变量名

# 显示图形
plt.show()



```
